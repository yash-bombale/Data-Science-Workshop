### Bivariate Analysis

```python
x = np.array([3,13,19,31,38,45,58,65,71,88])
y = np.array([11,16,24,37,41,58,56,78,81,95])
z = np.array([81,80,79,63,54,32,29,26,18,14])
```

```python
plt.title('My first plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.grid()
plt.xlim(0,100)
plt.ylim(0,100)
plt.xticks(range(0,101,10))
plt.yticks(range(0,101,10))
plt.plot(x, y)
```

```python
plt.title('My first plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.grid()
plt.xlim(0,100)
plt.ylim(0,100)
plt.xticks(range(0,101,10))
plt.yticks(range(0,101,10))
plt.plot(x, z)
```

```python
plt.title('My first plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.grid()
#markers: . o x v > ^ < D *
plt.plot(x, y, color='red', marker='o', ls=':', mfc='green')
```

```python
plt.title('My first plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.grid()
plt.plot(x, y, color='red', label = '2023')
plt.plot(x, z, color='blue', label = '2022')
plt.legend()
```

```python
plt.subplot(2,1,1) #(row,column,plot-no)
plt.plot(x,y)
plt.subplot(2,1,2)
plt.plot(x,z)
#plt.plot(x, y, color='red', marker='o', ls=':', mfc='green')
#plt.plot(x, z, color='blue', marker='o', ls=':', mfc='red')
```

```python
plt.subplot(1,2,1) #(row,column,plot-no)
plt.plot(x,y)
plt.subplot(1,2,2)
plt.plot(x,z)
```

```python
plt.figure(figsize=(16,9))
plt.title('My first plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.grid()
plt.plot(x, z, lw=5, marker = 'o', ms=20, mfc='y')
# ms - marker size
# mfc - marker face color
plt.savefig('myplot.png')
```

### Bar Plot

```python
enrollment = [104,112,256,178,126,180]
courses = ['C', 'Java', 'Python', 'R', 'PHP', 'JavaScript']
```

```python
plt.xlabel('Courses')
plt.ylabel('Enrolled Students')
plt.grid()
plt.bar(courses, enrollment, color='teal')
```

```python
plt.ylabel('Courses')
plt.xlabel('Enrolled Students')
plt.grid()
plt.barh(courses, enrollment, color='teal')
```

### Pie Chart

```python
plt.figure(figsize=(10,10))
plt.pie(enrollment, labels=courses, autopct='%2.2f', shadow=True);
```

### Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Salary_Data.csv')
```

```python
plt.scatter(df['YearsExperience'], df['Salary'])
```

### Seperate the input and output data

```python
x = df['YearsExperience'].values
y = df['Salary']
```

```python
x
```

```python
y;
```

```python
# input data should be in 2 dimension
x = x.reshape(30,1)
```

```python
x
```

### Import the linear regression class

```python
from sklearn.linear_model import LinearRegression
```

### Create the object of linear regression

```python
regressor = LinearRegression()
```

### Train the algorithm

```python
regressor.fit(x, y)
```

### Predict on the unseen data

```python
regressor.predict([[15]])
```

```python
regressor.predict([[20]])
```

```python
regressor.predict([[10]])
```

```python
regressor.score(x, y)
```
